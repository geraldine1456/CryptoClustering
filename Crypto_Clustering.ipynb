```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
# Import the module
from sklearn.preprocessing import StandardScaler

# Scaling the numeric columns
market_data_scaled = StandardScaler().fit_transform(df_market_data)

# Review the scaled data
market_data_scaled [:5]
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(market_data_scaled, columns=df_market_data.columns)

# Copy the crypto names from the original DataFrame
crypto_names = df_market_data.index.tolist()

# Set the coin_id column as index
df_market_data_scaled.index = crypto_names
df_market_data_scaled.index.name = "coin_id"

# Display the scaled DataFrame
df_market_data_scaled.head(5)
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_original = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df_market_data_scaled)
    inertia_original.append(k_model.inertia_)
    
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia_original": inertia_original}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia_original", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:**  4

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
scaled_clusters = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
scaled_clusters
```

```python
# Create a copy of the scaled DataFrame
df_market_data_scaled_predictions=df_market_data_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_market_data_scaled_predictions["CryptoClusters"]= scaled_clusters

# Display the copy of the scaled DataFrame
df_market_data_scaled_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
plot_original= df_market_data_scaled_predictions.hvplot.scatter(
        x="price_change_percentage_24h",
        y="price_change_percentage_7d",
        by="CryptoClusters",
        hover_cols = ["coin_id"],
        title = "Scatter Plot by Cryptocurrency Segment - k=4"
)
plot_original
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
market_data_scaled_pca  = pca.fit_transform(df_market_data_scaled)

# View the scaled PCA data
market_data_scaled_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 0.89503166 or approximately 89.5%

```python
# Create a new DataFrame with the PCA data.
df_market_data_scaled_pca = pd.DataFrame(market_data_scaled_pca, columns=["PC1", "PC2", "PC3"])

# Copy the crypto names from the original scaled DataFrame
crypto_names = df_market_data_scaled.index.tolist()

# Set the coin_id column as index
df_market_data_scaled_pca.index = crypto_names
df_market_data_scaled_pca.index.name = "coin_id"

# Display the scaled PCA DataFrame
df_market_data_scaled_pca.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=42)
    k_model.fit(df_market_data_scaled_pca)
    inertia_pca.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k, "inertia_pca": inertia_pca}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia_pca", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?
  **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?
  **Answer:** The best k value found is the same as when using the original data.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=42)
```

```python
# Fit the K-Means model using the PCA data
model.fit(df_market_data_scaled_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
scaled_PCA_clusters=model.predict(df_market_data_scaled_pca)

# Print the resulting array of cluster values.
scaled_PCA_clusters
```

```python
# Create a copy of the scaled PCA DataFrame
df_market_data_scaled_pca_predictions=df_market_data_scaled_pca.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_market_data_scaled_pca_predictions["CryptoClusters"]= scaled_PCA_clusters

# Display the copy of the scaled PCA DataFrame
df_market_data_scaled_pca_predictions.head()
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
plot_pca=df_market_data_scaled_pca_predictions.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="CryptoClusters",
    hover_cols = ["coin_id"],
    title = "Scatter Plot by Cryptocurrency Segment"
)
plot_pca
```

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
# Create a DataFrame for hvPlot visualization
elbow_df = pd.DataFrame({
    "k": k,
    "Inertia (Original)": inertia_original,
    "Inertia (PCA)": inertia_pca
})
elbow_df.hvplot(
    x="k", 
    y=["Inertia (Original)", "Inertia (PCA)"], 
    title="Elbow Method Comparison: Original vs. PCA Data",
    xlabel="K",
    ylabel="Inertia",
    line_width=2,
    legend=True
)

```

```python
# Composite plot to contrast the clusters
composite_plot = plot_original + plot_pca
composite_plot
```

**Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

**Answer:**  Using fewer features in K-Means clustering has a direct impact on the quality and interpretability of the clusters. 

* **Clusters are less distinct** - Overlapping clusters indicate that fewer features may limit how well cryptocurrencies can be separated.
   
* **Loss of variability** - Reduced features mean the dataset might not fully capture price movement trends, leading to less precise clustering.
   
* **Potential Over-Simplification** - While PCA helps reduce dimensionality, it may compress key patterns, making some distinctions harder to see.
